```python
from src import plot_acc, plot_loss, plot_params
import matplotlib.pyplot as plt
import pickle

plt.style.use('seaborn-darkgrid')
```

```python
with open('pickles/rgb_results_v2.pickle', 'rb') as f:
    rgb = pickle.load(f)
    
with open('pickles/audio_results_v2.pickle', 'rb') as f:
    audio = pickle.load(f)
```

```python
audio_gru_legends = ['128, 128, 128 units', '128, 256, 256 units', '128, 256, 512 units']
audio_lstm_legends = ['128, 128, 128 units', '128, 256, 256 units', '256, 256, 256 units']
audio_cnn_legends = ['128, 128, 128 units', '128, 256, 512 units']
audio_dnn_legends = ['256 units', '512 units', '1024 units', '256, 512 units', '512, 1024 units']

rgb_gru_legends = ['128, 128, 128 units', '128, 256, 256 units', '256, 256, 256 units']
rgb_lstm_legends = ['256, 256, 256 units', '256, 256, 512 units', '256, 512, 512 units']
rgb_cnn_legends = ['128, 128, 128 units', '256, 512, 512 units']
rgb_dnn_legends = ['256 units', '512 units', '1024 units', '256, 512 units', '512, 1024 units']
```

# RGB Features

Firstly, let's analyze the models based on RGB features.

## DNN Models

```python
plot_loss(rgb['losses'], sel='DNN', legends=rgb_dnn_legends, loc='upper right')
```

From validation loss graph, it can be seen that some models are overfitting after 10 epochs. 

```python
plot_acc(rgb['accs'], sel='DNN', legends=rgb_dnn_legends)
```

Following final models are chosen:
- 256 units
- 512 units
- 1024 units

## CNN Models

```python
plot_loss(rgb['losses'], sel='CNN', legends=rgb_cnn_legends, loc='upper right')
```

By looking at the loss functions, we can see that it decreases loss better than straightforward dnn models. 

```python
plot_acc(rgb['accs'], sel='CNN', legends=rgb_cnn_legends)
```

We see that somehow models performs badly on Hit@1 metric compared to Hit@5. Looking at the graphs two models selected for further tuning:
- 128, 128, 128 units
- 256, 512, 512 units

## LSTM Models

```python
plot_loss(rgb['losses'], sel='LSTM', legends=rgb_lstm_legends, loc='upper right')
```

By looking at the loss we can see again that all models performs in similar manner. To eliminate some of them, let's see about the acc parameters

```python
plot_acc(rgb['accs'], sel='LSTM', legends=rgb_lstm_legends)
```

As it can be seen, LSTM is not performing very well. No more training needed.

## GRU Models

```python
plot_loss(rgb['losses'], sel='GRU', legends=rgb_gru_legends, loc='upper right')
```

```python
plot_acc(rgb['accs'], sel='GRU', legends=rgb_gru_legends)
```
