```python
from src import plot_acc, plot_loss, plot_params
import matplotlib.pyplot as plt
import pickle

plt.style.use('seaborn-darkgrid')
```

```python
with open('pickles/rgb_results_v3.pickle', 'rb') as f:
    rgb = pickle.load(f)
    
with open('pickles/audio_results_v3.pickle', 'rb') as f:
    audio = pickle.load(f)
```

```python
audio_cnn_legends = ['128, 128, 128 units', '128, 256, 512 units']
audio_dnn_legends = ['256 units', '512 units', '1024 units', '256, 512 units', '512, 1024 units']

rgb_cnn_legends = ['128, 128, 128 units', '256, 512, 512 units']
rgb_dnn_legends = ['256 units', '512 units', '1024 units', '256, 512 units', '512, 1024 units']
```

**Note about the training process. It was seen that in about 10-15 mins of training model reaches to its maximum potential without even completing the first epoch. Training further didn't help with the performance. This shows us that we need to build more complex models to explore underlying features. This will be the next step to combine both audio and rgb features together to build a better model**
